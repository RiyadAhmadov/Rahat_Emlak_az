```python
# Create an empty DataFrame
df = pd.DataFrame(columns=['Location', 'Room', 'Currency', 'Area', 'Floor', 'Price'])

for i in linkler:
    Location = i.find(class_='display-flow-root color-black').text
    Room, Area, Floor = i.find(class_='color-black content-wrap-text').text.split(',')
    Currency = i.find(class_='color-black bolder content-wrap-text').text
    Price = i.find(class_='caption-bottom').text

    # Append the extracted values to the DataFrame
    df.loc[len(df)] = [Location, Room, Currency, Area, Floor, Price]
```

```python
# Create an empty DataFrame
df = pd.DataFrame(columns=['City' , 'Village', 'Room', 'New/Old', 'Area(m²)', 'Floor_Number)','Floor_From', 'Document', 'href_link','Price(AZN)'])
for i in range(1,2): 
    url = f'https://rahatemlak.az/alqi-satqi/menziller?page={i}'  
    response = requests.get(url)
    my_soup = BeautifulSoup(response.content)
    my_soup.prettify()
    linkler = my_soup.find_all('div', class_='item-box')
    for i in linkler:
        Location = i.find(class_='display-flow-root color-black').text
        Location = Location.replace('\n', '')
        Village = Location.split(', ')[1]
        City = Location.split(',')[0]
        if City == 'Xırdalan':
            Village = 'Xırdalan'
        Room, Area, Floor = i.find(class_='color-black content-wrap-text').text.split(',')
        Room = Room.replace(' otaqlı', '')
        Floor = Floor.replace(' mərtəbə', '')
        Area = Area.replace(' m²', '')
        Floor_Number  =  Floor.split('/')[0]
        Floor_From = Floor.split('/')[1]
        new_old = i.find(class_='color-black bolder content-wrap-text').text.split(' ')[0]
        Price = i.find(class_='caption-bottom').text
        Price = Price.replace('\n', '') 
        Price = Price.replace(' AZN', '') 
        
        i_tag_ipoteka = soup.find('i', class_='fa fa-percent')
        Document = i_tag_ipoteka['title']
        
        a_tag = i.find('a', href=True)
        href_link = a_tag['href']
        url = href_link 
        response = requests.get(url)
        my_soup = BeautifulSoup(response.content)
        links = my_soup.find_all('a', class_='button button-s button-round-small bg-green2-dark')
        for a in links:
            a_tag = a.find('a', href=True)
            href_link1 = a_tag['href']

        
        df.loc[len(df)] = [City , Village, Room, new_old, Area, Floor_Number, Floor_From, Document,href_link1 ,Price]
```
